# Data Analysis

- Remove Outliers from Dataset using boxplot method

```python
# find outliers using boxplot
df_rows = 3
df_cols = 5

fig, ax = plt.subplots(nrows = df_rows, ncols=df_cols, figsize = (20,10) )
index=0
ax= ax.flatten()
for col,value in df.items():
    sns.boxplot(data = df, y= col, ax = ax[index])
    index += 1
plt.tight_layout(pad = 0.5,w_pad =0.7 , h_pad =5)
```

```python
df.items
```

# Removal of outliers

```python
for key, value in df.items():
    q1 = value.quantile(0.25)
    q3 = value.quantile(0.75)
    iqr = q3 - q1
    ul = q3 + 1.5*iqr
    ll = q1 - 1.5*iqr
    
    value_col = value[(value <= ll) | (value >= ul)]
    perct = np.shape(value_col)[0] * 100 / np.shape(df)[0]
    print("Column %s outliers = %2f%%" % (key, perct))
    
```

- So from above correlation matrix to we identify that, MEDV feature more coreelatd with INDUS, NOX, TAX, PTRATIO, LSTAT, RM features.

Now we apply different threshold value of correlation coefficient and check which features gives us best prediction of price 

```python
def getCorrelatedFeatures(corrdata, threshold):
    df_features = []
    df_value = []
    for i, index in enumerate(corrdata.index):
        if abs(corrdata[index] > threshold):
            df_features.append(index)
            df_value.append(corrdata[index])
    
    df = pd.DataFrame(data = df_value, index = df_features, columns = ['CorrValue'])
    return df
```

```python
threshold = 0.3
corr_value = getCorrelatedFeatures(df_corr['Price'], threshold)
print(corr_value)
```

Above we can see we get same output as correlation matrix.

## Split target and response variable

```python
x = df.drop(['Price'], axis = 1)
y = df['Price']
```

## Data train test split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

## Linear regression

#### Training the model

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
# model training
lm.fit(x_train, y_train)
```

```python
# y intercept i.e b in y = m* x + b
lm.intercept_
```

```python
lm.coef_
```

```python
#Converting the coefficient values to a dataframe
coeffcients = pd.DataFrame([x_train.columns,lm.coef_]).T
coeffcients = coeffcients.rename(columns={0: 'Attribute', 1: 'Coefficients'})
coeffcients
```

### Model evaluation

```python
y_pred = lm.predict(x_train)
```

```python
from sklearn import metrics

print(f"R^2: {metrics.r2_score(y_train, y_pred)}")
print(f"Adjusted R^2: {1 - (1-metrics.r2_score(y_train, y_pred))*(len(y_train)-1)/(len(y_train)-x_train.shape[1]-1)}")
print(f"MSE: {metrics.mean_squared_error(y_train, y_pred)}")
```

#### Visualizing the differences between actual prices and predicted values

```python
plt.scatter(y_train, y_pred)
plt.xlabel("Price")
plt.ylabel("Predicted price")
plt.title("Price Vs Predicted price")
plt.show()
```

### Checking for linearity assumption

#### Checking residuals 

```python
plt.scatter(y_pred, y_train-y_pred)
plt.xlabel("Predicted price")
plt.ylabel("Residuals")
plt.title("Predicted Vs residuals")
plt.show()
```

There is no pattern visible in this plot and values are distributed equally around zero. So Linearity assumption is satisfied

### Checking for normality assumptions

```python
# using histogram 


plt.hist(y_train-y_pred)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Distribution of residuals")
plt.show()
```

Here the residuals are normally distributed. So normality assumption is satisfied

```python
# using Q-Q plot

import statsmodels.api as sm
fig = sm.qqplot(y_train-y_pred, line ='45')
plt.show()
```

Here, the points on this plot clearly don’t fall along the red line, so we would not assume that residuals are normally distributed.

```python
y_test_pred = lm.predict(x_test)
```

```python
from sklearn import metrics

print(f"R^2: {metrics.r2_score(y_test, y_test_pred)}")
print(f"Adjusted R^2: {1 - (1-metrics.r2_score(y_test, y_test_pred))*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)}")
print(f"MSE: {metrics.mean_squared_error(y_train, y_pred)}")
```

Here the model evaluations scores are almost matching with that of train data. So the model is not overfitting.
